```python
import pandas as pd
import matplotlib.pyplot as plt
```

This is a visualization of the amount of wall time spent by the computer in calculating each time step vs. the corresponding domain time. Some time steps take longer than others, either due to chemistry calls or writing the domain state to the history netcdf files

```python
data= pd.read_csv('core_timing.csv', header=0)
data['Domain_Time'] = pd.to_datetime(data['Domain_Time'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 4))
ax.plot(data['Domain_Time'], data['Duration(s)'], lw='.3')
ax.set_yscale('log')
ax.set_ylabel('Wall Time (s)')
ax.set_xlabel('Domain Time')
ax.set_xlim(pd.to_datetime('2023-03-20 09:00:00'), pd.to_datetime('2023-03-20 15:00:00'))
#ax.set_xlim(pd.to_datetime('2023-03-20 09:00:00'), pd.to_datetime('2023-03-20 12:00:00'))
```
